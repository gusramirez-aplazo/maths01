```python
from sympy import *

x = symbols('x')
f = 1 / x

plot(f)
```

```python
from sympy import *

x = symbols('x')
f = 1 / x
result = limit(f, x, oo)

print(result)
```

```python
from sympy import *

n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)

print(f'result not evaluated: {result}')
print(f'result evaluated: {result.evalf()}')
```
